```python
#Let's categorize gender as female, male and children
def male_female_child(passenger):
    age, Sex = passenger
    if age < 16:
        return 'child'
    else:
        return Sex
```

```python
#then we can apply the function to the data fram
titi['Person']=titi[['Age','Sex']].apply(male_female_child, axis=1)
```
